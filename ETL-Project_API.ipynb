```python
# Dependencies
import requests
from pprint import pprint
import pandas as pd

pd.options.mode.chained_assignment = None
```

```python
def get_price(price_url, price_date):
    date_url = price_url\
        + "?from=" + price_date\
        + "&to=" + price_date\
        + "&serietype=line"
    date_json = requests.get(date_url).json()
    
    try:
        prices_list = date_json["historical"]
        date_close = prices_list[0]["close"]
    except:
        date_close = None
    
    return date_close
```

```python
init_csv = "Data_after_cleaning.csv"
init_df = pd.read_csv(init_csv, index_col=0)
init_df = init_df.rename(columns={"LastSale": "Last_Sale"})
```

```python
nyse_df = init_df.dropna()
```

```python
base_url = "https://financialmodelingprep.com/api/v3/historical-price-full/"
start_date = "2018-12-31"

test_df = nyse_df
ytd_list = []

for index, row in test_df.iterrows():
    symbol_url = base_url + row["Symbol"]
    start_year_price = get_price(symbol_url, start_date)
    ytd_list.append(start_year_price)
    
last_sale_loc = test_df.columns.get_loc("Last_Sale")
test_df.insert(last_sale_loc + 1, "Start_of_Year_Sale", ytd_list)

test_df
```

```python
percent_series = ((test_df["Last_Sale"] - test_df["Start_of_Year_Sale"]) / test_df["Last_Sale"]) * 100

year_loc = test_df.columns.get_loc("Start_of_Year_Sale")
test_df.insert(year_loc + 1, "YTD_Change_Percent", percent_series)
test_df
```

```python
out_csv = "NYSE_Sector_Performance.csv"
test_df = test_df.dropna()
test_df = test_df.set_index("Symbol")
test_df = test_df.sort_index()
test_df.head()
test_df.to_csv(out_csv)
```
